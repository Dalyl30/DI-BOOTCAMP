```python
#Exercice 1
import scipy as sp
sp.__version__


#Exercice 2
#calculez la moyenne, la médiane, la variance et l'écart type à l'aide de SciPy


from scipy import stats

data = [12, 15, 13, 12, 18, 20, 22, 21]

# Moyenne
moyenne = stats.tmean(data)

# Médiane
mediane = stats.scoreatpercentile(data, 50)

# Variance
variance = stats.tvar(data)

# Écart-type
ecart_type = stats.tstd(data)

print("Moyenne :", moyenne)
print("Médiane :", mediane)
print("Variance :", variance)
print("Écart-type :", ecart_type)


```

```python
#Exercice3

#Générer une distribution normale à l'aide de SciPy avec une moyenne de 50 et un écart type de 10.
#Représenter graphiquement cette distribution.


import matplotlib.pyplot as plt
from scipy.stats import norm
#Paramètres de la distribution
moyenne=50
ecart_type=10
#Génération
x=range(20,81)
#Calcul de la densité de probabilité
y = norm.pdf(x, loc=moyenne, scale=ecart_type)
#Traçé de la courbe
plt.plot(x,y)
plt.title("Distribution normale (μ=50, σ=10) ")
plt.xlabel("Valeurs")
plt.ylabel("Densité de probabilité")
plt.grid(True)

plt.show


#Exercice 4

#Effectuer un test T sur deux ensembles de données générées aléatoirement
import numpy as np
from scipy import stats

# Génération des données
data1 = np.random.normal(50, 10, 100)
data2 = np.random.normal(60, 10, 100)

# Test t de Student
t_stat, p_value = stats.ttest_ind(data1, data2)

print("Statistique t :", t_stat)
print("P-value :", p_value)

# Interprétation
if p_value < 0.05:
    print("Différence significative")
else:
    print("Pas de différence significative")

```
